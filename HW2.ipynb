### Задача 2 (10 баллов)

Вы участвуете в организации эксперимента и пишете программу, которая принимает на вход данные i-го объекта. Известно, что форматом хранения данных является таблица, где строка соответствует наблюдению (в данном случае — уникальному объекту), а столбец — переменной. Следовательно, в ячейке ij находится значение j-ой переменной (показателя) для i-го наблюдения (объекта). Напишите программу, которая будет принимать на вход по одной такой строке. Программа должна:

- принимать на вход разделитель (для элементов строки). Разделителем может быть `','`, `';'` и `' '` — *1 балл*;
- принимать на вход названия колонок (переменных) через указанный разделитель. Гарантируется, что разделитель всегда будет корректным — *1 балл*;
- принимать на вход типы колонок (переменных) через указанный разделитель; возможные типы — `int`, `float`, `bool`, `str` — *1 балл*;
- принимать на вход строки до тех пор, пока на вход не будет подано значение `'TRANSMISSION STOPPED'` — *2 балла*;
- если количество элементов в строке не соответствует заявленному ранее, печатать сообщение `'INVALID ELEMENTS NUMBER'` и начинать следующий ввод — *1 балл*;
- возвращать и печатать для пользователя словарь, в котором ключами будут названия, а значениями — списки со значениями каждой переменной для каждого объекта. Значения переменной должны быть приведены к соответствующему типу. Гарантируется, что входная спецификация типов корректна — *4 балла*.

**Примечание.** Использование модуля `pandas` <u>запрещено</u>.

**Пример:**

```
input:

Введите разделитель: ;
Введите названия колонок: height;weight;gender
Введите типы данных: float;float;str
190;84;male
160;50;female
175;65;male
183;100;male
TRANSMISSION STOPPED

output:

{'height': [190.0, 160.0, 175.0, 183.0],
 'weight': [84.0, 50.0, 65.0, 100.0],
 'gender': ['male', 'female', 'male', 'male']}
```

**Критерии:**

См. в условии.

````python
# запрашиваем у пользователя входные данные:
delim = input('Введите разделитель: ') 
if delim not in [',',';',' ']:
    raise ValueError("Invalid delimiter!")
col_names = input('Введите названия колонок: ')
data_types = input('Введите типы данных: ') 

def add_obs():
    global delim
    # превратить входные данные в списки
    colnames = col_names.split(delim)
    types = data_types.split(delim)
    # проверяем, соответствует ли количество указанных типов количеству колонок
    if len(colnames) != len(types):
        raise ValueError("The number of entered data types does not correspond to the number of columns")
    # проверяем, есть ли недопустимые типы данных среди указанных
    if not(set(types).issubset(set(['int', 'float', 'bool', 'str']))):
        raise ValueError("Invalid data type!")
# создаём функцию, осуществляющую форматирование данных, введённых пользователем, в соответствии с заданными форматами      
    def format_fn(inp):
        nonlocal types
        global delim
        strobs = inp.split(delim)
        # обрабатываем случай, когда количество введенных данных не соответствует количеству колонок
        if len(strobs) != len(types):
            print("INVALID ELEMENTS NUMBER")
            obs = None
        else:    
            obs = []
            for ind in range(len(types)):
                fmt = types[ind].replace('"', "").replace("'","") ## maybe one can omit this part
                obs.append(eval(fmt)(strobs[ind]))
        return obs     

# создаём функцию, которая будет принимать произвольное количество наблюдений по заданному формату, пока не встретит ключевую фразу  
    def get_data():
        lst = []
        while True:
        # запрашиваем у пользователя данные
            obs = input()
        # условие выхода
            if obs == "TRANSMISSION STOPPED":
                break
            else:
                ## форматируем данные
                obs = format_fn(obs)
                # если количество полей в введённых данных и количество колонок не совпало, требуем новый ввод
                if obs == None:
                    continue
                else:
                    lst.append(obs)

                # создаём словарь, если нет
                dictionary = {}
                cols = []
                for ind, elem in enumerate(colnames):
                    tmplist = []
                    for i in range(len(lst)):
                        tmplist.append(lst[i][ind])
                    cols.append(tmplist)
                    # заполняем словарь ключами из списка названий колонок и значений из сформированного выше списка
                    dictionary[elem] = cols[ind]

        return dictionary    
                
    return get_data
    
print(add_obs()())   
````
