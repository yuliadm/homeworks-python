### Пример
**Условие.** Напишите программу, которая просит пользователя ввести электронную почту и проверяет корректность её введения. По условию, почтовый адрес является корректным, если в нём присутствуют один и только один символ `@` (при этом слева от этого символа должен находиться хотя бы ещё один [исключая `@`]). _Это не полный список критериев, но для учебного примера достаточно этих ограничений._ При нарушении хотя бы одного условия программа должна ответить, что такая почта не принимается, и разъяснить причину. Если почта корректна, об этом также надо сообщить пользователю.

**Алгоритм решения:**
1. Получить на вход почту в виде строки (функция `input`).
2. Проверить разные условия (условная конструкция и подходящие операторы):
 - проверить наличие ровно одного символа `@` (строковый метод `.count()`) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`);
 - проверить наличие символов перед `@` (строковый метод `.startswith()` — если начинается с `@`, то условие не выполнено) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`).
3. Сообщить пользователю, что почта корректна, если это так (функция `print()`).

**Решение:**

````python
email = input()  # шаг 1

if email.count('@') != 1:  # шаг 2-1
    print('Введённый почтовый адрес некорректен. Он должен содержать ровно одну @')
elif email.startswith('@'):  # шаг 2-2
    print('Введённый почтовый адрес некорректен. Он должен иметь хотя бы один символ перед @')
else:  # шаг 3
    print('Поздравляем, ваша почта принята!')
````

### Задача 1 (5 баллов)

Пользователь подаёт на вход программе действительные числа через пробел. Он привык пользоваться Excel, поэтому разделяет числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять» — в этом случае слово надо игнорировать). **Ваша задача** — вывести минимум и максимум последовательности через запятую с пробелом. Гарантируется, что будет введено хотя бы одно число.

**Пример:**

```
input:
-2.56 25,002 восемь

output:
-2.56, 25.002
```

**Критерии:**

- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание присутствует и соответствует решению — 5 баллов;
- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание отсутствует / задача решена удовлетворительно (учтено одно из условий), описание присутствует и соответствует решению — 4 балла;
- задача решена с серьёзными ошибками, но описание присутствует и соответствует решению — 3 балла;
- иное — 0 баллов.

**Алгоритм решения:**

1. Получить на вход последовательность действительных чисел в виде строки (`input`).
2. Заменить в строке запятые, отделяющие количество разрядов (если они есть), точками (метод `.replace()`).
3. Конвертировать строку в список (метод `.split()`).
4. Объявить объекты minimum и maximum (здесь будут жить соответственно минимальное и максимальное значения введённой последовательности).
4. Для каждого элемента в списке попробовать конвертировать значение в числовой формат `float` (`try-except`) и сделать следущее: 
    - если операция успешна, и minimum, maximum пусты, положить значение в minimum, maximum;
    - если операция успешна и minimum не пуст, сравнить текущее числовое значение со значением в minimum, если текущее значение меньше значения, лежащего в minimum, заменить значение в minimum на текущее,
    - если операция успешна и maximum не пуст, сравнить текущее числовое значение со значением в maximum, если текущее значение больше значения, лежащего в maximum, заменить значение в maximum на текущее,
    - иначе (это рассчитано на случай ввода словами) выдать предупреждение и проигнорировать нечисловое значение.
5. Вывести минимум и максимум последовательности через запятую.

````python
# test input:
# numbers = "-2.56 25,002 восемь -3,456 567 seven 0.5 6,758 three,dogs,jump.over.the,fence 100.04 -4,059.05 -54.55"

# Inquire the sequence from the user:
numbers = input("Enter a space delimited sequence of real numbers:")

# replace all commas with dots and convert to list
numbers = numbers.replace(',','.').split()

# create placeholders for the min and max values 
minimum = maximum = []
# cycle through all elements in the list and try to convert them to float
for elem in numbers:
    try:
        flt = float(elem) 
        # print the value if it is a number
        print(flt)
    # if ValueError is encountered, print the warning and ignore    
    except ValueError:
        print("Incorrect input. This value will be disregarded.")
    # if the placeholder is empty and the value can be converted to a number, put the number into the placeholder    
    if not minimum and not maximum and flt:
        minimum = maximum = flt
    # if some value is already in the placeholder, compare the current number with 
    # the number in the placeholder and replace it if the current value is smaller (larger)  
    elif (flt and flt < minimum):
        minimum = flt
    elif (flt and flt > maximum):
        maximum = flt
    # if the value is not a number, ignore 
    else:
        pass    
# print the minimum:    
print('{}, {}'.format(minimum, maximum))          
````

### Задача 2 (5 баллов)

Вы хотите помочь пользователю из первой задачи. Представим, что он хочет округлить действительное число, но не знает, как сделать это в Python. Зато он помнит, как это можно сделать в Excel: с помощью функций [ОКРУГЛ](https://support.microsoft.com/ru-ru/office/%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F-%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%BB-c018c5d8-40fb-4053-90b1-b3e7f61a213c), [ОКРУГЛВВЕРХ](https://support.microsoft.com/ru-ru/office/%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%BB%D0%B2%D0%B2%D0%B5%D1%80%D1%85-%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F-%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%BB%D0%B2%D0%B2%D0%B5%D1%80%D1%85-f8bc9b23-e795-47db-8703-db171d0c42a7) и [ОКРУГЛВНИЗ](https://support.microsoft.com/ru-ru/office/%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%BB%D0%B2%D0%BD%D0%B8%D0%B7-%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F-%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%BB%D0%B2%D0%BD%D0%B8%D0%B7-2ec94c73-241f-4b01-8c6f-17e6d7968f53). Пользователь будет вводить Excel-выражение (любое из возможных, согласно документации по ссылкам), а вам необходимо преобразовать его в соответствующий Python-код и вывести результат.

**Пример:**

```
input:
ОКРУГЛ(1,5136;2)

output:
1.51
```

**Критерии:**

- задача решена корректно (полностью реализован маппинг Excel-Python для всех трёх функций), описание присутствует и соответствует решению — 5 баллов;
- задача решена корректно (полностью реализован маппинг Excel-Python для всех трёх функций), описание отсутствует / задача решена удовлетворительно (реализован маппинг Excel-Python для двух из трёх функций), описание присутствует и соответствует решению — 4 балла;
- решение неполное (реализован маппинг Excel-Python для одной из трёх функций), описание присутствует и соответствует решению / решение неполное (реализован маппинг Excel-Python для двух из трёх функций, но с ошибками), описание присутствует и соответствует решению — 3 балла;
- иное — 0 баллов.

**Алгоритм решения:**

1. Получить на вход команду EXCEL в виде строки (`input`).
2. Вернуть подстроку от скобки до скобки по индексам (определяются с помощью метода `.find()`), заменить в представлении числв запятую на точку (replace()) и разделить значения на список (split()) 
3. Использовать факт, что оператор // возвращает целое с округлением вниз (так не надо прописывать случаи, когда второй аргумент (количество разрядов) отрицательный, положиткльный и т.п.):
    - если в введённой строке используется `ОКРУГЛВНИЗ` (т.е. начинается подстрокой "ОКРУГЛВН"), выполнить округление как `k//10**(-p)*10**(-p)`, где k - число, которое надо округлить, p - количество разрядов[^1];
    - если в введённой строке используется `ОКРУГЛВВЕРХ` (т.е. начинается подстрокой "ОКРУГЛВВ"), выполнить округление как `-(-k//10**(-p)*10**(-p))`;
    - в иных случаях (`ОКРУГЛ` и возможные ошибки) в названии функции, просто использовать функцию `round()`;
4. Вывести результат округления.

[^1]: Здесь и в следующем пункте можно поставить костыль: `round(k//10**(-p)*10**(-p),p)`,  потому что возникает погрешность в точности при работе с `float`, если необходимо округлить до 4 и более знаков после запятой. 
Этот видно, если запустить пример `excelcmd = 'ОКРУГЛВВЕРХ(-176,56734;4)'`.

````python
# test input
# excelcmd = 'ОКРУГЛВВЕРХ(-176,56734;4)'

# Inquire the excel expression from the user:
excelcmd = input("Enter the excel command for rounding:")

# subset the string from the opening bracket to the closing bracket (using indices) and split to obtain a list 
beg = excelcmd.find('(')
end = excelcmd.find(')')
inp = excelcmd[beg+1:end].replace(',', '.').split(';')

k = float(inp[0]) # number to round (base)
p = int(inp[1])   # number of decimals (precision)

# we will make use of the following hack: Python integer division returns the floor
# if cmd == 'ОКРУГЛВНИЗ()'
if excelcmd.startswith('ОКРУГЛВН'):
    # calculate floor using //
    r = k//10**(-p)*10**(-p) 
# if cmd == 'ОКРУГЛВВЕРХ()'    
elif excelcmd.startswith('ОКРУГЛВВ'):
    # calculate ceiling using //
    r = -(-k//10**(-p)*10**(-p))
# if cmd == 'ОКРУГЛ()' и др.    
else:
    # classical rounding
    r = round(k, p) 
# print the rounded value
print(r)
````
